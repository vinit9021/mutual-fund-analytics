```python
from pathlib import Path
import warnings

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.stats import linregress
from IPython.display import display

warnings.filterwarnings("ignore")

# ---------------------------------------------------------
# Paths
# ---------------------------------------------------------

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()

EDA_DIR = PROJECT_ROOT / "data" / "processed" / "eda"
RAW_DIR = PROJECT_ROOT / "data" / "raw"
PERFORMANCE_DIR = PROJECT_ROOT / "data" / "processed" / "performance"
CHART_DIR = PROJECT_ROOT / "reports" / "charts"

PERFORMANCE_DIR.mkdir(parents=True, exist_ok=True)
CHART_DIR.mkdir(parents=True, exist_ok=True)

print("Project root:", PROJECT_ROOT)
print("EDA folder:", EDA_DIR)
print("Performance output folder:", PERFORMANCE_DIR)
print("Charts folder:", CHART_DIR)


# ---------------------------------------------------------
# Load NAV and scheme master data
# ---------------------------------------------------------

nav = pd.read_csv(EDA_DIR / "eda_nav_40_schemes.csv")
scheme_master = pd.read_csv(EDA_DIR / "eda_scheme_master_40.csv")

nav["date"] = pd.to_datetime(nav["date"])
nav = nav.sort_values(["amfi_code", "date"]).reset_index(drop=True)

print("NAV shape:", nav.shape)
print("Scheme master shape:", scheme_master.shape)

display(nav.head())
display(scheme_master.head())


# ---------------------------------------------------------
# Basic validation
# ---------------------------------------------------------

required_nav_cols = {"amfi_code", "scheme_name", "fund_house", "category", "date", "nav"}
missing_cols = required_nav_cols - set(nav.columns)

if missing_cols:
    raise ValueError(f"Missing columns in NAV file: {missing_cols}")

if nav["nav"].isna().any():
    print("Warning: Missing NAV values found. Forward filling within each scheme.")
    nav["nav"] = nav.groupby("amfi_code")["nav"].ffill()

if (nav["nav"] <= 0).any():
    raise ValueError("Invalid NAV found. NAV should be greater than 0.")

print("NAV validation completed.")


# ---------------------------------------------------------
# 1. Compute daily returns
# daily_return = NAV_t / NAV_t-1 - 1
# ---------------------------------------------------------

nav["daily_return"] = nav.groupby("amfi_code")["nav"].pct_change()

daily_returns = nav.dropna(subset=["daily_return"]).copy()

daily_return_distribution = (
    daily_returns.groupby(["amfi_code", "scheme_name"])
    .agg(
        observations=("daily_return", "count"),
        mean_daily_return=("daily_return", "mean"),
        std_daily_return=("daily_return", "std"),
        min_daily_return=("daily_return", "min"),
        max_daily_return=("daily_return", "max"),
        median_daily_return=("daily_return", "median"),
        skew_daily_return=("daily_return", "skew"),
    )
    .reset_index()
)

daily_return_distribution.to_csv(
    PERFORMANCE_DIR / "daily_return_distribution.csv",
    index=False
)

print("Daily return distribution saved.")
display(daily_return_distribution.head())


# ---------------------------------------------------------
# Validate daily return distribution
# ---------------------------------------------------------

overall_daily_stats = daily_returns["daily_return"].describe(percentiles=[0.01, 0.05, 0.25, 0.5, 0.75, 0.95, 0.99])

print("Overall daily return distribution:")
display(overall_daily_stats)

extreme_returns = daily_returns[
    (daily_returns["daily_return"] > 0.15) |
    (daily_returns["daily_return"] < -0.15)
]

print("Extreme daily return rows beyond +/-15%:", len(extreme_returns))

if len(extreme_returns) > 0:
    display(extreme_returns[["date", "amfi_code", "scheme_name", "nav", "daily_return"]].head(20))


# ---------------------------------------------------------
# Helper function for CAGR
# CAGR = (NAV_end / NAV_start) ^ (1/n) - 1
# ---------------------------------------------------------

def compute_cagr_for_years(group, years):
    group = group.sort_values("date").copy()

    end_date = group["date"].max()
    end_nav = group.loc[group["date"] == end_date, "nav"].iloc[-1]

    target_start_date = end_date - pd.DateOffset(years=years)

    available_start_date = group["date"].min()

    if available_start_date > target_start_date:
        # Full period not available. Use earliest available date and annualize.
        start_date = available_start_date
        period_available = False
    else:
        # Use the closest date on or after target start date.
        possible_rows = group[group["date"] >= target_start_date]
        start_date = possible_rows["date"].min()
        period_available = True

    start_nav = group.loc[group["date"] == start_date, "nav"].iloc[0]

    actual_years = (end_date - start_date).days / 365.25

    if actual_years <= 0:
        return np.nan, start_date, end_date, actual_years, period_available

    cagr = (end_nav / start_nav) ** (1 / actual_years) - 1

    return cagr, start_date, end_date, actual_years, period_available


# ---------------------------------------------------------
# 2. Compute CAGR for 1yr, 3yr, 5yr
# ---------------------------------------------------------

cagr_rows = []

for amfi_code, group in nav.groupby("amfi_code"):
    scheme_name = group["scheme_name"].iloc[0]
    fund_house = group["fund_house"].iloc[0]
    category = group["category"].iloc[0]

    row = {
        "amfi_code": amfi_code,
        "scheme_name": scheme_name,
        "fund_house": fund_house,
        "category": category,
    }

    for years in [1, 3, 5]:
        cagr, start_date, end_date, actual_years, period_available = compute_cagr_for_years(group, years)

        row[f"cagr_{years}yr"] = cagr
        row[f"cagr_{years}yr_percent"] = cagr * 100
        row[f"cagr_{years}yr_start_date"] = start_date
        row[f"cagr_{years}yr_end_date"] = end_date
        row[f"cagr_{years}yr_years_used"] = actual_years
        row[f"cagr_{years}yr_full_period_available"] = period_available

    cagr_rows.append(row)

cagr_comparison = pd.DataFrame(cagr_rows)

cagr_comparison.to_csv(
    PERFORMANCE_DIR / "cagr_comparison.csv",
    index=False
)

print("CAGR comparison table saved.")
display(
    cagr_comparison[
        [
            "amfi_code",
            "scheme_name",
            "category",
            "cagr_1yr_percent",
            "cagr_3yr_percent",
            "cagr_5yr_percent",
            "cagr_5yr_years_used",
            "cagr_5yr_full_period_available",
        ]
    ].head()
)


# ---------------------------------------------------------
# 3. Sharpe Ratio and Sortino Ratio
# Rf = 6.5% annual repo rate proxy
# ---------------------------------------------------------

RISK_FREE_RATE_ANNUAL = 0.065
TRADING_DAYS = 252

rf_daily = RISK_FREE_RATE_ANNUAL / TRADING_DAYS

risk_rows = []

for amfi_code, group in daily_returns.groupby("amfi_code"):
    scheme_name = group["scheme_name"].iloc[0]
    fund_house = group["fund_house"].iloc[0]
    category = group["category"].iloc[0]

    returns = group["daily_return"].dropna()

    mean_daily_return = returns.mean()
    std_daily_return = returns.std()

    downside_returns = returns[returns < 0]
    downside_std = downside_returns.std()

    annualized_return = mean_daily_return * TRADING_DAYS
    annualized_volatility = std_daily_return * np.sqrt(TRADING_DAYS)

    sharpe_ratio = np.nan
    sortino_ratio = np.nan

    if std_daily_return and std_daily_return > 0:
        sharpe_ratio = ((mean_daily_return - rf_daily) / std_daily_return) * np.sqrt(TRADING_DAYS)

    if downside_std and downside_std > 0:
        sortino_ratio = ((mean_daily_return - rf_daily) / downside_std) * np.sqrt(TRADING_DAYS)

    risk_rows.append({
        "amfi_code": amfi_code,
        "scheme_name": scheme_name,
        "fund_house": fund_house,
        "category": category,
        "mean_daily_return": mean_daily_return,
        "annualized_return": annualized_return,
        "std_daily_return": std_daily_return,
        "annualized_volatility": annualized_volatility,
        "sharpe_ratio": sharpe_ratio,
        "sortino_ratio": sortino_ratio,
    })

risk_metrics = pd.DataFrame(risk_rows)

risk_metrics.to_csv(
    PERFORMANCE_DIR / "risk_metrics_day4.csv",
    index=False
)

print("Sharpe and Sortino ratios saved.")
display(risk_metrics.sort_values("sharpe_ratio", ascending=False).head(10))


# ---------------------------------------------------------
# 4. Create / load benchmark data
# Nifty 50 and Nifty 100
# ---------------------------------------------------------
# If you later get real benchmark data, place it here:
# data/raw/benchmarks.csv
#
# Accepted formats:
# 1) date,nifty50,nifty100
# 2) date,nifty50_return,nifty100_return
#
# If no benchmark file exists, this code creates a proxy benchmark
# from the current 40-scheme universe.

benchmark_file = RAW_DIR / "benchmarks.csv"

return_pivot = daily_returns.pivot_table(
    index="date",
    columns="amfi_code",
    values="daily_return"
).sort_index()

if benchmark_file.exists():
    benchmarks = pd.read_csv(benchmark_file)
    benchmarks["date"] = pd.to_datetime(benchmarks["date"])
    benchmarks = benchmarks.sort_values("date")

    if {"nifty50", "nifty100"}.issubset(benchmarks.columns):
        benchmarks["nifty50_return"] = benchmarks["nifty50"].pct_change()
        benchmarks["nifty100_return"] = benchmarks["nifty100"].pct_change()

    elif {"nifty50_return", "nifty100_return"}.issubset(benchmarks.columns):
        benchmarks["nifty50"] = (1 + benchmarks["nifty50_return"].fillna(0)).cumprod() * 100
        benchmarks["nifty100"] = (1 + benchmarks["nifty100_return"].fillna(0)).cumprod() * 100

    else:
        raise ValueError("benchmarks.csv must contain either nifty50/nifty100 or nifty50_return/nifty100_return columns.")

    benchmarks = benchmarks[["date", "nifty50", "nifty100", "nifty50_return", "nifty100_return"]]

    print("Real benchmark file loaded:", benchmark_file)

else:
    print("No real benchmark file found. Creating benchmark proxy from EDA fund universe.")

    large_cap_codes = nav[nav["category"].str.contains("Large", case=False, na=False)]["amfi_code"].unique().tolist()

    if len(large_cap_codes) == 0:
        large_cap_codes = nav["amfi_code"].drop_duplicates().head(10).tolist()

    large_cap_returns = return_pivot[large_cap_codes].mean(axis=1)
    all_fund_returns = return_pivot.mean(axis=1)

    benchmark_returns = pd.DataFrame(index=return_pivot.index)
    benchmark_returns["nifty50_return"] = (0.90 * large_cap_returns) + (0.10 * all_fund_returns)
    benchmark_returns["nifty100_return"] = (0.75 * large_cap_returns) + (0.25 * all_fund_returns)

    benchmark_returns = benchmark_returns.dropna().reset_index()

    benchmark_returns["nifty50"] = (1 + benchmark_returns["nifty50_return"]).cumprod() * 100
    benchmark_returns["nifty100"] = (1 + benchmark_returns["nifty100_return"]).cumprod() * 100

    benchmarks = benchmark_returns[["date", "nifty50", "nifty100", "nifty50_return", "nifty100_return"]]

benchmarks.to_csv(
    PERFORMANCE_DIR / "benchmark_returns.csv",
    index=False
)

print("Benchmark data saved.")
display(benchmarks.head())


# ---------------------------------------------------------
# 5. Alpha and Beta using OLS regression
# fund returns on Nifty 100 returns
# scipy.stats.linregress
# Alpha = intercept * 252
# ---------------------------------------------------------

alpha_beta_rows = []

benchmark_returns = benchmarks[["date", "nifty100_return"]].dropna()

for amfi_code, group in daily_returns.groupby("amfi_code"):
    scheme_name = group["scheme_name"].iloc[0]
    fund_house = group["fund_house"].iloc[0]
    category = group["category"].iloc[0]

    merged = pd.merge(
        group[["date", "daily_return"]],
        benchmark_returns,
        on="date",
        how="inner"
    ).dropna()

    if len(merged) < 30:
        beta = np.nan
        alpha_daily = np.nan
        alpha_annual = np.nan
        r_value = np.nan
        p_value = np.nan
        stderr = np.nan
        observations = len(merged)
    else:
        result = linregress(
            merged["nifty100_return"],
            merged["daily_return"]
        )

        beta = result.slope
        alpha_daily = result.intercept
        alpha_annual = alpha_daily * TRADING_DAYS
        r_value = result.rvalue
        p_value = result.pvalue
        stderr = result.stderr
        observations = len(merged)

    alpha_beta_rows.append({
        "amfi_code": amfi_code,
        "scheme_name": scheme_name,
        "fund_house": fund_house,
        "category": category,
        "alpha_daily": alpha_daily,
        "alpha_annual": alpha_annual,
        "alpha_annual_percent": alpha_annual * 100 if pd.notna(alpha_annual) else np.nan,
        "beta": beta,
        "r_squared": r_value ** 2 if pd.notna(r_value) else np.nan,
        "p_value": p_value,
        "stderr": stderr,
        "observations": observations,
    })

alpha_beta = pd.DataFrame(alpha_beta_rows)

alpha_beta.to_csv(
    PERFORMANCE_DIR / "alpha_beta.csv",
    index=False
)

print("Alpha and beta table saved.")
display(alpha_beta.sort_values("alpha_annual", ascending=False).head(10))


# ---------------------------------------------------------
# 6. Maximum Drawdown
# max drawdown = min(NAV / running_max - 1)
# Also find peak date, trough date, and recovery date
# ---------------------------------------------------------

drawdown_rows = []

for amfi_code, group in nav.groupby("amfi_code"):
    group = group.sort_values("date").copy()

    scheme_name = group["scheme_name"].iloc[0]
    fund_house = group["fund_house"].iloc[0]
    category = group["category"].iloc[0]

    group["running_max"] = group["nav"].cummax()
    group["drawdown"] = group["nav"] / group["running_max"] - 1

    trough_idx = group["drawdown"].idxmin()
    trough_date = group.loc[trough_idx, "date"]
    trough_nav = group.loc[trough_idx, "nav"]
    max_drawdown = group.loc[trough_idx, "drawdown"]

    before_trough = group[group["date"] <= trough_date].copy()
    peak_nav = before_trough["nav"].max()
    peak_date = before_trough.loc[before_trough["nav"].idxmax(), "date"]

    after_trough = group[group["date"] > trough_date].copy()
    recovery_rows = after_trough[after_trough["nav"] >= peak_nav]

    if len(recovery_rows) > 0:
        recovery_date = recovery_rows["date"].iloc[0]
        recovered = True
        days_to_recover = (recovery_date - trough_date).days
    else:
        recovery_date = pd.NaT
        recovered = False
        days_to_recover = np.nan

    drawdown_rows.append({
        "amfi_code": amfi_code,
        "scheme_name": scheme_name,
        "fund_house": fund_house,
        "category": category,
        "max_drawdown": max_drawdown,
        "max_drawdown_percent": max_drawdown * 100,
        "peak_date": peak_date,
        "trough_date": trough_date,
        "recovery_date": recovery_date,
        "recovered": recovered,
        "days_to_recover": days_to_recover,
        "peak_nav": peak_nav,
        "trough_nav": trough_nav,
    })

drawdown = pd.DataFrame(drawdown_rows)

drawdown.to_csv(
    PERFORMANCE_DIR / "max_drawdown.csv",
    index=False
)

print("Maximum drawdown table saved.")
display(drawdown.sort_values("max_drawdown").head(10))


# ---------------------------------------------------------
# 7. Expense ratio data
# For scorecard, expense ratio is required.
# If scheme_master already has expense_ratio_percent, use it.
# Otherwise create controlled category-based values.
# ---------------------------------------------------------

expense = scheme_master.copy()

if "expense_ratio_percent" not in expense.columns:
    print("expense_ratio_percent not found. Creating category-based estimated expense ratios.")

    category_expense_map = {
        "Large Cap": 0.85,
        "Mid Cap": 0.95,
        "Small Cap": 1.05,
        "Flexi Cap": 0.90,
        "ELSS": 0.95,
        "Index Fund": 0.25,
        "Hybrid": 0.80,
        "Sectoral": 1.10,
    }

    def assign_expense_ratio(category):
        category = str(category)

        for key, value in category_expense_map.items():
            if key.lower() in category.lower():
                return value

        return 0.90

    expense["expense_ratio_percent"] = expense["category"].apply(assign_expense_ratio)

expense = expense[["amfi_code", "expense_ratio_percent"]].drop_duplicates()

expense.to_csv(
    PERFORMANCE_DIR / "expense_ratio_used.csv",
    index=False
)

display(expense.head())


# ---------------------------------------------------------
# 8. Fund Scorecard
# Composite:
# 30% x 3yr return rank
# 25% x Sharpe rank
# 20% x Alpha rank
# 15% x expense ratio rank inverse
# 10% x max DD rank inverse
# ---------------------------------------------------------

scorecard = (
    cagr_comparison[
        [
            "amfi_code",
            "scheme_name",
            "fund_house",
            "category",
            "cagr_1yr_percent",
            "cagr_3yr_percent",
            "cagr_5yr_percent",
        ]
    ]
    .merge(
        risk_metrics[
            [
                "amfi_code",
                "annualized_return",
                "annualized_volatility",
                "sharpe_ratio",
                "sortino_ratio",
            ]
        ],
        on="amfi_code",
        how="left"
    )
    .merge(
        alpha_beta[
            [
                "amfi_code",
                "alpha_annual_percent",
                "beta",
                "r_squared",
            ]
        ],
        on="amfi_code",
        how="left"
    )
    .merge(
        drawdown[
            [
                "amfi_code",
                "max_drawdown_percent",
                "peak_date",
                "trough_date",
                "recovery_date",
                "days_to_recover",
            ]
        ],
        on="amfi_code",
        how="left"
    )
    .merge(
        expense,
        on="amfi_code",
        how="left"
    )
)

# Percentile ranks
# Higher is better: 3yr CAGR, Sharpe, Alpha
# Lower is better: expense ratio
# Less negative max drawdown is better, so higher max_drawdown_percent is better.

scorecard["rank_3yr_return"] = scorecard["cagr_3yr_percent"].rank(pct=True)
scorecard["rank_sharpe"] = scorecard["sharpe_ratio"].rank(pct=True)
scorecard["rank_alpha"] = scorecard["alpha_annual_percent"].rank(pct=True)
scorecard["rank_expense_inverse"] = scorecard["expense_ratio_percent"].rank(ascending=False, pct=True)
scorecard["rank_max_dd_inverse"] = scorecard["max_drawdown_percent"].rank(pct=True)

scorecard["fund_score"] = (
    0.30 * scorecard["rank_3yr_return"] +
    0.25 * scorecard["rank_sharpe"] +
    0.20 * scorecard["rank_alpha"] +
    0.15 * scorecard["rank_expense_inverse"] +
    0.10 * scorecard["rank_max_dd_inverse"]
) * 100

scorecard["fund_score"] = scorecard["fund_score"].round(2)

scorecard = scorecard.sort_values("fund_score", ascending=False).reset_index(drop=True)
scorecard["overall_rank"] = np.arange(1, len(scorecard) + 1)

scorecard.to_csv(
    PERFORMANCE_DIR / "fund_scorecard.csv",
    index=False
)

print("Fund scorecard saved.")
display(scorecard.head(10))


# ---------------------------------------------------------
# 9. Benchmark comparison chart
# Top 5 funds vs Nifty 50 and Nifty 100 over 3 years
# Tracking error = std(fund_return - benchmark_return) x sqrt(252)
# ---------------------------------------------------------

top5_codes = scorecard.head(5)["amfi_code"].tolist()

latest_date = nav["date"].max()
start_3yr_date = latest_date - pd.DateOffset(years=3)

top5_nav = nav[
    (nav["amfi_code"].isin(top5_codes)) &
    (nav["date"] >= start_3yr_date)
].copy()

benchmark_3yr = benchmarks[benchmarks["date"] >= start_3yr_date].copy()

# Normalise fund NAV to base 100
top5_nav["base_nav"] = top5_nav.groupby("amfi_code")["nav"].transform("first")
top5_nav["normalised_nav"] = top5_nav["nav"] / top5_nav["base_nav"] * 100

benchmark_3yr["nifty50_normalised"] = benchmark_3yr["nifty50"] / benchmark_3yr["nifty50"].iloc[0] * 100
benchmark_3yr["nifty100_normalised"] = benchmark_3yr["nifty100"] / benchmark_3yr["nifty100"].iloc[0] * 100

plt.figure(figsize=(16, 8))

for amfi_code in top5_codes:
    fund_data = top5_nav[top5_nav["amfi_code"] == amfi_code].copy()
    fund_name = fund_data["scheme_name"].iloc[0]

    short_name = (
        fund_name.replace(" - Direct Plan Growth Plan - Growth Option", "")
        .replace(" - Growth Option - Direct Plan", "")
        .replace(" FUND-DIRECT PLAN -GROWTH", "")
        .replace(" - Direct Plan - Growth", "")
        .replace(" - Growth - Direct", "")
        .replace(" Direct Growth", "")
        .replace(" Fund", "")
    )

    plt.plot(
        fund_data["date"],
        fund_data["normalised_nav"],
        linewidth=2,
        label=short_name[:45]
    )

plt.plot(
    benchmark_3yr["date"],
    benchmark_3yr["nifty50_normalised"],
    linewidth=3,
    linestyle="--",
    label="Nifty 50"
)

plt.plot(
    benchmark_3yr["date"],
    benchmark_3yr["nifty100_normalised"],
    linewidth=3,
    linestyle="--",
    label="Nifty 100"
)

plt.title("Top 5 Funds vs Nifty 50 and Nifty 100, Last 3 Years")
plt.xlabel("Date")
plt.ylabel("Normalised Value, Base = 100")
plt.legend(loc="best", fontsize=9)
plt.grid(True, alpha=0.3)
plt.tight_layout()

benchmark_chart_path = CHART_DIR / "20_benchmark_comparison_top5.png"

plt.savefig(
    benchmark_chart_path,
    dpi=300,
    bbox_inches="tight"
)

plt.show()

print("Benchmark comparison chart saved:", benchmark_chart_path)


# ---------------------------------------------------------
# 10. Tracking error
# ---------------------------------------------------------

tracking_error_rows = []

fund_returns_3yr = daily_returns[
    (daily_returns["amfi_code"].isin(top5_codes)) &
    (daily_returns["date"] >= start_3yr_date)
].copy()

benchmark_returns_3yr = benchmarks[
    benchmarks["date"] >= start_3yr_date
][["date", "nifty50_return", "nifty100_return"]].copy()

for amfi_code in top5_codes:
    fund_data = fund_returns_3yr[fund_returns_3yr["amfi_code"] == amfi_code].copy()

    scheme_name = fund_data["scheme_name"].iloc[0]

    merged = pd.merge(
        fund_data[["date", "daily_return"]],
        benchmark_returns_3yr,
        on="date",
        how="inner"
    ).dropna()

    tracking_error_nifty50 = (
        (merged["daily_return"] - merged["nifty50_return"]).std() *
        np.sqrt(TRADING_DAYS)
    )

    tracking_error_nifty100 = (
        (merged["daily_return"] - merged["nifty100_return"]).std() *
        np.sqrt(TRADING_DAYS)
    )

    tracking_error_rows.append({
        "amfi_code": amfi_code,
        "scheme_name": scheme_name,
        "tracking_error_nifty50": tracking_error_nifty50,
        "tracking_error_nifty50_percent": tracking_error_nifty50 * 100,
        "tracking_error_nifty100": tracking_error_nifty100,
        "tracking_error_nifty100_percent": tracking_error_nifty100 * 100,
        "observations": len(merged),
    })

tracking_error = pd.DataFrame(tracking_error_rows)

tracking_error.to_csv(
    PERFORMANCE_DIR / "tracking_error.csv",
    index=False
)

print("Tracking error table saved.")
display(tracking_error)


# ---------------------------------------------------------
# 11. Final output summary
# ---------------------------------------------------------

print("DAY 4 PERFORMANCE ANALYTICS COMPLETED SUCCESSFULLY")
print()
print("Created files:")
print("-", PERFORMANCE_DIR / "daily_return_distribution.csv")
print("-", PERFORMANCE_DIR / "cagr_comparison.csv")
print("-", PERFORMANCE_DIR / "risk_metrics_day4.csv")
print("-", PERFORMANCE_DIR / "alpha_beta.csv")
print("-", PERFORMANCE_DIR / "max_drawdown.csv")
print("-", PERFORMANCE_DIR / "expense_ratio_used.csv")
print("-", PERFORMANCE_DIR / "fund_scorecard.csv")
print("-", PERFORMANCE_DIR / "benchmark_returns.csv")
print("-", PERFORMANCE_DIR / "tracking_error.csv")
print("-", CHART_DIR / "20_benchmark_comparison_top5.png")
```
